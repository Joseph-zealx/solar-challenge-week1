```python
sierra = pd.read_csv('data/sierra_leone_clean.csv')
```

```python
togo = pd.read_csv('data/togo_clean.csv')
```

```python
benin['Country'] = 'Benin'
```

```python
sierra['Country'] = 'Sierra Leone'
```

```python
togo['Country'] = 'Togo'
```

```python
df_all = pd.concat([benin, sierra, togo])
```

```python
sns.boxplot(x='Country', y='GHI', data=df_all), plt.title("GHI Distribution by Country"), plt.show()
```

```python
from scipy.stats import f_oneway
```

```python
f_stat, p_val = f_oneway(benin['GHI'], sierra['GHI'], togo['GHI'])
print(f"ANOVA p-value: {p_val}")
```
